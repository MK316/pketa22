```python
#@markdown üê¨ Choose language & dialect:

def tts(mytext):
  text_to_say = mytext
  language_to_choose = "en" #@param ["en", "ko", "fr", 'es']
  language = language_to_choose

  sp = "0" #@param [0, 1]
  sp = int(sp)
  speed = bool(sp)

  dialect = "com" #@param ["com","co.uk","com.au","ca","co.in","ie"]
  print("Play language accent: %s"%language_to_choose)
  print("Play local accent: %s"%dialect)
  language = language_to_choose

  gtts_object = gTTS(text = text_to_say,
                     lang = language, tld = dialect,
                    slow = speed)  
  gtts_object.save("mytext.wav")
  return Audio("mytext.wav")

# text_to_say = text

# intro_text = "I'll repeat a word three times. "
# text_to_say = intro_text + text_to_say 

# tts(text_to_say)
```
